```python
# create new columns
import requests
trip_data['geocode_data_start'] = ''
trip_data['address_start']=''

# function that handles the geocoding requests

# def reverseGeocode(latlng):
#     result = {}
#     url = 'https://maps.googleapis.com/maps/api/geocode/json?latlng={0}&key={1}'
#     apikey = 'XXX'
#     request = url.format(latlng, apikey)
#     data = json.loads(requests.get(request).text)
#     if len(data['results']) > 0:
#         result = data['results'][0]
#     return result
def reverseGeocode(latlng):
    result = {}
    url = 'https://maps.googleapis.com/maps/api/geocode/json?latlng={}'
    request = url.format(latlng)
    data = requests.get(request).json()
    if len(data['results']) > 0:
        result = data['results'][0]
    return result

for i, row in trip_data[:3].iterrows():
# for i, row in [0,1,2]:
    trip_data['geocode_data_start'][i] = reverseGeocode(trip_data['start station latitude'][i].astype(str) + ',' + trip_data['start station longitude'][i].astype(str))

for i, row in trip_data[:3].iterrows():
    if 'address_components' in row['geocode_data_start']:
        for component in row['geocode_data_start']['address_components']:
             trip_data['address_start'] = row['geocode_data_start']['formatted_address']
            
```

## Daily customers travel much more slowly than subscribers
#### Look at the plot


```python
trip_max_by_user = trip_data.groupby(['usertype'])['tripduration'].max()
trip_max_by_user.sort_values(inplace=True)
# type(labour_data_by_specialization)
plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title('tripduration for daily customers and subscribers')
plt.xlabel('Type of users')
plt.ylabel('Trip Duration')
trip_max_by_user.plot(rot=45)
plt.show()
```

## Visitors love to make trips during morning : 8 and 9 AM and in the evening 5 and 6 PM.

```python
times = pd.DatetimeIndex(trip_data.starttime)
group_by_hour = trip_data.groupby([times.hour]).count()
fig,ax = plt.subplots()
ax.set_xlabel("Day Hours",fontsize=12)
ax.set_ylabel("No. of visitors",fontsize=12)
group_by_hour.plot(ax=ax,kind='bar',title ="No. of visitors/frequency vs. hours",figsize=(10,6), legend=False )
```

# Most of the people love to go out Tuesday and Wednesday
###  Remember the day of the week with Monday=0, Sunday=6

```python
trip_data['weekday'] = trip_data['starttime'].apply(lambda x: x.weekday())
weekdays_only = trip_data['weekday'] < 5
weekdays_only_visitors = trip_data.groupby(['weekday']).count()
fig,ax = plt.subplots()
ax.set_xlabel("Day Hours",fontsize=12)
ax.set_ylabel("No. of visitors",fontsize=12)
weekdays_only_visitors.plot(ax=ax,kind='bar',title ="No. of visitors/frequency vs. hours",figsize=(10,6), legend=False )
```

# Let us suppose 1 == Male and 2 == Female. The graph shows that Male are taking long to complete the trip than Female groups

```python

trip_data_gender = trip_data[trip_data.gender != 0]
trip_max_by_gender = trip_data_gender.groupby(['gender'])['tripduration'].max()
trip_max_by_gender.sort_values(inplace=True)
# type(labour_data_by_specialization)
plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title('Male and Female tripduration plot')
plt.xlabel('Male/Female')
plt.ylabel('Trip Duration')
trip_max_by_gender.plot(rot=45, kind='bar')
plt.show()
```

```python
import datetime
def get_age(data):
    return datetime.datetime.today().year - int(data)
    
```

```python
trip_data_age = trip_data[trip_data['birth year'] != '\N']
trip_data_age['age'] = trip_data_age['birth year'].apply(get_age)
```

```python
trip_data_age.head()
```

# People in age range 30 to 80 are making the trip. 30 Age groups are making longer trips

```python
trip_max_by_age = trip_data_age.groupby(['age'])['tripduration'].max()
trip_max_by_age.sort_values(inplace=True)
# type(labour_data_by_specialization)
plt.figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title('tripduration for different age groups')
plt.xlabel('Ages')
plt.ylabel('Trip Duration')
trip_max_by_age.plot(rot=45, kind='bar')
plt.show()
```
