### In this example, we will examine a house predicting problem.

#### Load house using pandas

```python
import pandas as pd
```

```python
house_data = pd.read_csv('data/house_data.csv')
```

```python
house_data.head()
```

#### Let's visuallize the data and see if there exists any relationships between response variable (price) and explanatory variable (sqft_living)

```python
%matplotlib inline
import matplotlib.pylab as plt

plt.scatter(house_data['price'], house_data['sqft_living'])
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.title('sqft_living and price Relationship')
plt.show()
```
